```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters())
```

```python
#This will not automatically work
#Upload best_quadcam.pt to sessions storage
#It's in the github
#Maybe change the path if you put it in a different place

model = torch.load("../best_quadcam.pt")

model = model["model"] 
```

```python
temp = ''
p = 0.2

for module in model.model.modules():

  print(count_parameters(module))
  
  if isinstance(module, nn.Conv2d): #Prune the Conv2D layers
    print(module)
    print("\n")
    temp = module.float()
    prune.l1_unstructured(temp, name="weight", amount=p) #Zero out 20% of the connections

if (p < 1):
  print("The model is " + str(p*100) + "% smaller!")
else:
  print("The model has " + str(p) + "fewer elements!")
```
